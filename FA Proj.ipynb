```python
app_data = pd.read_csv('IS453 Group Assignment - Application Data.csv')
app_data.head()
```

```python
# Data Preparation

# Check for missing data for application data (more than 4 values missing)
rows_w_gt_3_na = app_data[app_data.isnull().sum(axis=1) > 3].shape[0]
print('Percent of total rows missing more than 3 values: ' + str("{:.1%}".format(rows_w_gt_3_na/app_data.shape[0])))
```

```python

# Check for missing data for bureau data (more than 4 values missing)
rows_w_gt_3_na = bureau_data[bureau_data.isnull().sum(axis=1) > 3].shape[0]
print('Percent of total rows missing more than 3 values: ' + str("{:.1%}".format(rows_w_gt_3_na/bureau_data.shape[0])))
```

```python
# Check percent of missing col (application data)

col_w_na = pd.DataFrame((app_data.isnull().sum().sort_values(ascending=False)/app_data.shape[0])).applymap("{0:.0%}".format)

print(col_w_na[0][:30])
```

```python
# Check percent of missing col (application data)

col_w_na = pd.DataFrame((bureau_data.isnull().sum().sort_values(ascending=False)/bureau_data.shape[0])).applymap("{0:.0%}".format)

print(col_w_na[0][:20])
```
